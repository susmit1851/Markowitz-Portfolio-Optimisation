```python
import os
import pandas as pd
import numpy as np

df=pd.DataFrame()
var_df=pd.DataFrame()
variances=[]
for i in range(len(os.listdir('data'))):
    files=os.listdir('data')[i]
    filename=os.path.join('data', files)
    df1=pd.read_csv(filename)
    df1=df1.drop([0,1])
    close=np.array(list(df1['Close']), dtype=np.float64)
    returns=[]
    for i in range(1, len(close)):
        ret=(close[i]-close[i-1])/close[i-1]
        returns.append(ret)
    df[f'col_{i}']=returns
    var_val=np.var(np.array(close, dtype=np.float64))
    variances.append(var_val)
var_df['Volatility']=variances
```

```python
from utils import *

num_securities=10
returns=pd.read_csv('Data.csv')
returns=returns[[f'col_{i}' for i in range(10)]]

sigma=np.array(returns.cov())
mu=np.array(returns.mean())

gamma=0.05
G=np.linalg.cholesky(sigma)

with Model('Markowitz_Optimisation') as M:
    x=M.variable("x", num_securities, mf.Domain.greaterThan(0.0))
    M.constraint('budget', mf.Expr.sum(x), mf.Domain.equalsTo(1.0))
    M.objective('obj', mf.ObjectiveSense.Maximize, mf.Expr.dot(mu, x))
    M.constraint('risk', mf.Expr.vstack(gamma, 0.5, mf.Expr.mul(G.T, x)), mf.Domain.inRotatedQCone())
    M.solve()
    expected_return=M.primalObjValue()
    portfolio=x.level()
    print("Optimal Portfolio Allocation:", portfolio)
    print("Expected Portfolio Return:", expected_return)
```

```python
securities = [f'Security {i+1}' for i in range(10) ]

plt.figure(figsize=(10, 6))
plt.bar(securities, portfolio, color='blue')
plt.xlabel('Securities')
plt.ylabel('Portfolio Allocation')
plt.title('Optimized Portfolio Allocation')
plt.show()

# ---- Plot Efficient Frontier ----
# Efficient frontier requires solving for different risk levels (gamma)
risk_levels = np.linspace(0.01, 0.2, 50)
returns = []

for gamma_val in risk_levels:
    with mf.Model('Markowitz_Optimisation') as M:
        x = M.variable("x", num_securities, mf.Domain.greaterThan(0.0))
        M.constraint('budget', mf.Expr.sum(x), mf.Domain.equalsTo(1.0))
        M.objective('obj', mf.ObjectiveSense.Maximize, mf.Expr.dot(mu, x))
        M.constraint('risk', mf.Expr.vstack(gamma_val, 0.5, mf.Expr.mul(G.T, x)),
                     mf.Domain.inRotatedQCone())
        M.solve()
        returns.append(M.primalObjValue())
print(returns)
# Plot Efficient Frontier
plt.figure(figsize=(10, 6))
plt.plot(risk_levels, returns, marker='o', color='green')
plt.xlabel('Risk (Standard Deviation)')
plt.ylabel('Return')
plt.title('Efficient Frontier')
plt.grid(True)
plt.show()
```
